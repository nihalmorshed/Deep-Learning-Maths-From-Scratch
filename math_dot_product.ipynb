```python
import numpy as np
import torch as tr
```

```python
vec1 = np.array([1, 0, 2,5,-2])
vec2 = np.array([2, 8, -6, 1, 0])
```

using "np.dot()" function

```python
print(np.dot(vec1,vec2))
```

using basic multiply and add compuation

```python
print(np.sum(vec1*vec2))
```

Using PyTorch

```python
tnsr1 = tr.tensor([0,3,2])
tnsr2 = tr.tensor([1,0,6])
```

```python
print(tr.dot(tnsr1,tnsr2))
```

using basic multiply and add compuation

```python
print(tr.sum(tnsr1*tnsr2))
```

To compute the dot product of 2D, flatten the tensors first

```python
tnsr1 = tr.tensor([[0,3,2],[-3,-3,1],[1,0,2]])
tnsr2 = tr.tensor([[1,0,6],[2,-1,0],[5,1,4]])
```

```python
print(tr.dot(tnsr1.flatten(),tnsr2.flatten()))
```
