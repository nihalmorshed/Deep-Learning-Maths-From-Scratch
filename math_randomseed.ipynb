```python
import numpy as np
import torch 
```

```python
np.random.randn(5)
```

```python
np.random.seed(28)

print(np.random.randn(6))
print(np.random.randn(6))
# Rerunning this block will produce the same 6 numbers in each rows
```

### New seed mechanism in numpy

```python
randomseed1 = np.random.RandomState(17)
randonseed2 = np.random.RandomState(48532)
print("First line for randomseed1: ", randomseed1.randn(5))
print("First line for randomseed2: ",randonseed2.randn(5))
print("Second line for randomseed1: ", randomseed1.randn(5))
print("Second line for randomseed2: ",randonseed2.randn(5))


# Rerunning this block will produce the same 5 numbers in each rows
```

## Using PyTorch

```python
torch.randn(5)
```

```python
torch.manual_seed(17)
torch.randn(5)


# Rerunning this block will produce the same 5 numbers
```

PyTorch's "manual_seed" dont affect the numpy's random seed
