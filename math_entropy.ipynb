```python
import numpy as np
```

### wrong way to compute entropy

```python
p =0.25

entropy = - p* np.log(p)
```

```python
print(f"Wrong Value of Entropy: {entropy}")
```

Correct way to compute entropy

```python
x = [0.25, 0.75]

entropy1 = 0

for p in x:
    entropy1 += -p * np.log(p)

print(f"Correct Entropy: {entropy1}")
```

### Binary Cross Entropy

```python
## explicitly written out formula for entropy (Binary Cross Entropy)

h = -p * np.log(p) - (1-p) * np.log(1-p)
print(f"Correct Entropy: {h}")
```

# Cross-Entropy

```python
# All probability must sum to 1
p =[1,0]
q = [0.25,0.75]

h1 = 0

for i in range(len(p)):
    h1 += -p[i]*np.log(q[i])

print(f"CRoss Entropy: {h1}")
```

```python
## Binary cross entropy
h1 = -p[0] * np.log(q[0]) - p[1] * np.log(q[1])
print(f"Binary Cross Entropy: {h1}")
```

```python
# Simplified 
h1 = np.log(q[0])
print(f"Simplified Binary Cross Entropy: {h1}")
```
