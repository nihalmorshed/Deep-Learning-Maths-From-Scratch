```python
import scipy.stats as stats
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n1 = 30 # no. of samples in the dataset 1
n2 = 40 # no. of samples in the dataset 2

mew1 = 1 # assumed population mean for dataset 1
mew2 = 2 # assumed population mean for dataset 2
```

```python
# generating the data 
data1 = mew1 + np.random.randn(n1)
data2 = mew2 + np.random.randn(n2)
```

```python
# plot the data points
plt.plot(np.zeros(n1),data1,'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
```


1. **`np.zeros(n1)`** - The x-axis values. Creates an array of n1 zeros, so all points are plotted at x=0 (vertically aligned).

2. **`data1`** - The y-axis values. Your actual data points to be plotted.

3. **`'ro'`** - The marker style string:
   - `'r'` = red color
   - `'o'` = circle marker shape

4. **`markerfacecolor='w'`** - The fill color of the marker circles. `'w'` means white, so the circles are filled with white (creating a hollow/outlined appearance).

5. **`markersize=14`** - The size of the markers in points. This controls how large the circles appear on the plot.

Combined, this plots your `data1` points as red circles with white fill and a size of 14 points, all aligned vertically at x=0.


1. **`plt.xlim([-1,2])`** - Sets the x-axis limits:
   - `[-1, 2]` specifies the range of the x-axis display
   - The x-axis will display from -1 to 2
   - This creates padding around your data points (which are at x=0 and x=1) for better visualization

2. **`plt.xticks([0,1], labels=['Group 1', 'Group 2'])`** - Customizes the x-axis tick marks and labels:
   - `[0,1]` specifies where tick marks should appear on the x-axis (at positions 0 and 1)
   - `labels=['Group 1', 'Group 2']` replaces the numeric tick labels with descriptive text
   - Instead of showing "0" and "1", the axis will display "Group 1" and "Group 2"
   - This makes the plot more interpretable by labeling each data cluster with a meaningful name

Together, these commands create a plot where your two datasets are displayed as vertical columns of points at x=0 and x=1, with the x-axis labeled as "Group 1" and "Group 2" respectively.

```python
# ttest_ind -> independent samples t-test where data1 is the accuracies from deep learning model 1 and data2 is the accuracies from deep learning model 2  
t_value, p_value = stats.ttest_ind(data1, data2) 
```

```python
print(t_value)
print(p_value)
```

Based on the t-test output:

- **t-value: -4.079310729781239**
- **p-value: 0.00012089207967140169**

**Conclusion:**

There is a **statistically significant difference** between Group 1 and Group 2:

1. **p-value interpretation**: The p-value (≈ 0.00012) is much less than the typical significance level of 0.05 (α = 0.05), meaning we reject the null hypothesis.

2. **Null hypothesis rejection**: The null hypothesis (that both groups have the same population mean) is rejected. The data provides strong evidence that the two groups have different means.

3. **t-value interpretation**: The negative t-value (-4.08) indicates that Group 1 has a significantly lower mean than Group 2, which aligns with your setup where mew1 = 1 and mew2 = 2.

4. **Statistical significance**: With a p-value of 0.00012, there's only about 0.01% probability of observing this difference by chance if the groups truly had equal means. This is very strong evidence of a real difference between the groups.

In summary: **Group 1 and Group 2 have significantly different population means** at the 0.05 significance level.
