```python
import numpy as np
import torch 
import torch.nn as nn
import matplotlib.pyplot as plt
```

Manually compute the softmax function

```python
z = [2,3,5]
numerator = np.exp(z)
denom = np.sum(np.exp(z))
sigma = numerator/denom


```

```python
print(sigma)
print(np.sum(sigma))
```

```python
z = np.random.randint(low=-5, high=15, size=12)
numerator1 = np.exp(z)
denom1 = np.sum(np.exp(z))
sigma1 = numerator1/denom1
```

```python
plt.plot(z, sigma1, 'ko')
plt.xlabel("Softmax Input (output of a deep learning unit)")
plt.ylabel("Sofmax Output $\sigma$")
plt.yscale('linear')
# plt.yscale('log') #-> exponential in log scale looks linear
plt.title(f"$\sum\sigma $ = {np.sum(sigma1)}")


```

### Using Torch.nn 
We are going to create an instance of the softmax activation class from torch

```python
z1 = np.random.randint(-5,20, size= 25)
softFunc = nn.Softmax(dim=0)

```

### Now we apply that softmax func to this data

```python
sigmaTorch = softFunc(torch.Tensor(z1)) #conversion to Tensor is needed
print(sigmaTorch) # different datatype
```

```python
plt.plot(z1, sigmaTorch, 'ko')
plt.xlabel("Softmax Input (output of a deep learning unit)")
plt.ylabel("Sofmax Output $\sigma$")
# plt.yscale('linear')
# plt.yscale('log') #-> exponential in log scale looks linear
plt.title(f"$\sum\sigma $ = {torch.sum(sigmaTorch)}")


```
