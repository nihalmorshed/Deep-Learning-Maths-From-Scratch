```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.001,1,num=200) #Returns "num" evenly spaced samples, calculated over the interval [start, stop].
logx = np.log(x)
```

```python
fig = plt.figure(figsize=(10,4)) # defines t figure size in inches
```

```python
plt.plot(x,logx, "ks-", markerfacecolor='w') # k means black, s means square, - means line style and markerfacecolor='w'overrides the inside fill color of the markers to "white"
plt.xlabel("X")
plt.ylabel("LogX")
```

### demonstrate that natural exponent and log are inverse of each other

```python
x2 = np.linspace(0.001, 1, num=20)

log_x2 = np.log(x2)
exp_x = np.exp(x2)

```

```python
plt.plot(x2,x2, "r-")
plt.plot(x2, np.exp(log_x2), "o", markersize=8)
plt.plot(x2, np.log(exp_x), "x", markersize=8)

plt.xlabel("x")
plt.ylabel("f(g(x))")
plt.legend(["Unity", "exp(log_x2)", "log(exp_x)"])
```
