```python
import numpy as np
import torch as tr
```

```python
vctr = np.array([[1,47,354]])
print(vctr)
```

Use "np.transpose(vector)" to transpose

```python
transpose_vctr = np.transpose(vctr)
print(transpose_vctr)
```

Use ".T" to transpose

```python
transpose_vctr = vctr.T
print(transpose_vctr), print(' ')
```

Transposing for a Matrix

```python
matrx = np.array([[1,64,78,943],[74,35,59,37],[33,17,978,41]])
print(matrx)
```

```python
# transpose_mtrx = np.transpose(matrx)
transpose_mtrx = matrx.T

print(transpose_mtrx)
```

Transpose using PyTorch (vector & matrix)

```python
tnsr_1d = tr.tensor([[1,64,78,943]])
tnsr_2d = tr.tensor([[1,64,78,943],[74,35,59,37],[33,17,978,41]])

trnsps_1d = tnsr_1d.T
trnsps_2d = tnsr_2d.T
```

```python
print(trnsps_1d)
print(trnsps_2d)
```

lets examine the dataTypes

```python
print("The data type of vctr is:", type(vctr))
print("The data type of matrx is:", type(matrx))
print("The data type of tnsr_1d is:", type(tnsr_1d))
print("The data type of tnsr_2d is:", type(tnsr_2d))
```
