```python
import numpy as np
import torch as tr
```

Genrating Random array of given dimension

```python
r1 = np.random.randn(3,4)
r2 = np.random.randn(4,5)
r3 = np.random.randn(5,7)
r4 = np.random.randn(5,4)
```

```python
rounded_result1 = np.round(np.matmul(r1,r2),3)
print(rounded_result1)
```

```python
rounded_result2 = np.round(np.matmul(r1,r3),3)
print(rounded_result2)
```

```python
rounded_result3 = np.round(np.matmul(r2,r3),2)
print(rounded_result3)
```

```python
rounded_result5 = np.round(r1@r4,2)
print(rounded_result5)
```

```python
rounded_result4 = np.round(r1@r4.T,2)
print(rounded_result4)
```

Using PyTorch

```python
t1 = tr.randn(3,4)
t2 = tr.randn(4,5)
t3 = tr.randn(5,7)
t4 = tr.randn(5,4)
```

```python
# multiplying numpy array with pytorch tensor
print(t1@r2) 
```

### some times we need to covert numpy array to tensor to perform supported operations

```python
r2_tensor = tr.tensor(r2, dtype=tr.float)
print(r2_tensor@t3)
```
