```python
import numpy as np
import torch
import torch.nn as nn
```

```python
vect = np.array([1,40,2,-3])

max_val = np.max(vect)
min_val = np.min(vect)

arg_min = np.argmin(vect)
arg_max = np.argmax(vect)
```

```python
print(f"Max Value: {max_val}, Min value: {min_val}")
print(f"ArgMax Value: {arg_max}, ArgMin value: {arg_min}")
```

### With MAtrix

```python
M = np.array([
   [0,1,10],
   [20,8,5] 
])

print(M)
```

```python
minval1 = np.min(M) # Entire matrix min
minval2 = np.min(M, axis=0) # minimum values across rows (for each column)
minval3 = np.min(M, axis=1) # minimum values across columns (for each row)
```

```python
print(minval1)
print(minval2) # columns 
print(minval3) # rows
```

```python
argmin1 = np.argmin(M)
argmin2 = np.argmin(M, axis=0) # index of the minimum values across rows (for each column)
argmin3 = np.argmin(M, axis=1) # index of the minimum values across columns (for each row)
```

```python
print(argmin1)
print(argmin2) # column 
print(argmin3) # row
```

### USing PyTorch

```python
tsr = torch.Tensor([
    [0,1,10],
    [20,8,5] 
])
print(tsr)
```

```python
tsr_min1 = torch.min(tsr)
tsr_min2 = torch.min(tsr,axis= 0)
tsr_min3 = torch.min(tsr,axis= 1)
```

```python
print("Minimum within the array: ")
print(tsr_min1)
print("Minimum across the row (each column): ")
print(tsr_min2) # column 

print("Minimum across the column (each row): ")
print(tsr_min3) # row
```

```python
print(tsr_min2.values) 
print(tsr_min2.indices) 

print(tsr_min3.values) # row
print(tsr_min3.indices) # row
```
